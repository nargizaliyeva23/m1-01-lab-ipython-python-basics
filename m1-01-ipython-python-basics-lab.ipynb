#### Task 2: Reason about mutability and references
Create a list called values with at least three elements, assign it to alias, and append one new value using alias. Show the contents of both names after the change. Next, make a copy using list(values) or values.copy(), append a different value to the copy, and show the contents of both. Repeat the same pattern with a dictionary called record and an alias called record_alias, then a copied dictionary called record_copy. End the task with a short markdown paragraph explaining what changed and why, and how that affects passing objects into functions.

```python
values=[10,20,30]

alias=values 
alias.append(20)

print('Values: ' , values)
print('Alias: ',alias) 
```

```python
values_copy=values.copy()
values_copy.append(99)

print('Values: ',values)
print('values_copy: ',values_copy)
```

```python
record={"name":"Cons","Surname":"Smith"}

record_alias=record

record_alias["Age"]=45

print("record: " ,record)
print("record_alias: ",record_alias)
```

```python

record_copy=record.copy()
record_copy["Country"]="England"

print("record: ",record)
print("record_copy: ",record_copy)
```

When we assign a list or a dictionary to a new variable, both names point to the same object in memory. This means that changes made through the alias also affect the original object.
However, when we create a copy of the list or dictionary, changes to the copy do not affect the original. Only the copy itself will change.
This behavior is important for functions: if I pass a mutable object (list or dictionary) to a function and modify it, the original object will change unless I pass a copy. Using copies prevents unintended side effects.

#### Task 4: Apply control flow to a small record list
Create a list of at least 12 dictionaries representing user activity events with keys user_id, event_type, and duration_seconds. Include at least two records with invalid duration_seconds values (such as negative numbers or non-numeric strings). Write a loop that builds a new list called cleaned_events containing only valid records and adds a new key duration_minutes to each cleaned record. After the loop, verify two things: the count of cleaned_events matches the number of valid records you expect, and every cleaned record contains all required keys. Inspect two cleaned records in IPython to confirm the transformation.

```python
user_events = [
    {"user_id": 101, "event_type": "login", "duration_seconds": 45},
    {"user_id": 102, "event_type": "logout", "duration_seconds": 10},
    {"user_id": 103, "event_type": "video_watch", "duration_seconds": 300},
    {"user_id": 104, "event_type": "quiz_attempt", "duration_seconds": 120},
    {"user_id": 105, "event_type": "page_view", "duration_seconds": 30},
    {"user_id": 106, "event_type": "login", "duration_seconds": 60},
    {"user_id": 107, "event_type": "video_watch", "duration_seconds": "unknown"},  # invalid
    {"user_id": 108, "event_type": "quiz_attempt", "duration_seconds": 200},
    {"user_id": 109, "event_type": "page_view", "duration_seconds": 15},
    {"user_id": 110, "event_type": "logout", "duration_seconds": -5},  # invalid
    {"user_id": 111, "event_type": "video_watch", "duration_seconds": 400},
    {"user_id": 112, "event_type": "login", "duration_seconds": 50},
]

```

```python
cleaned_events=[]
for event in user_events:
    duration = event["duration_seconds"]
    if isinstance(duration,(int,float)) and duration>=0:
        new_event=event.copy()
        new_event["duration_minutes"]=duration/60
        cleaned_events.append(new_event)

        
print("count of valid events:",len(cleaned_events))

for event in cleaned_events:
    print(event)

print("\nTwo sample events:")
print(cleaned_events[0])
print(cleaned_events[1])
    
```

#### Task 5: Summarize and capture a small output
Compute the following summaries using only core Python structures. First, build a dictionary of event counts by event_type. Second, build a dictionary of average duration_minutes by event_type. Third, compute the number of unique users using a set. After computing the summaries, validate one result by recomputing it in a different way, such as summing the event counts and comparing to len(cleaned_events).

Next, build a CSV-formatted string with three columns: metric, key, and value. Include one row per event type for the counts, one row per event type for the averages, and a final row for the unique user count. If you choose to save the string to disk while working, read it back and confirm one value matches your in-memory results. Include the final CSV string in your submission file.

```python
event_count={}

for event in cleaned_events:
    etype=event["event_type"]
    if etype in event_count:
        event_count[etype] += 1
    else:
        event_count[etype] = 1
print(f'Event counts by event_type : {event_count}')
```

```python
duration_sum={}
duration_count={}

for event in cleaned_events:
    etype=event["event_type"]
    minutes=event["duration_minutes"]

    if etype in duration_sum:
        duration_sum[etype]+=minutes
        duration_count[etype]+=1
    else:
        duration_sum[etype] = minutes
        duration_count[etype] = 1


average_duration={}
for etype in duration_sum:
    average_duration[etype]=duration_sum[etype]/duration_count[etype]

print (f"Average duration time : {average_duration}")
```

```python
#unique user
unique_users=set()

for event in cleaned_events:
    unique_users.add(event["user_id"])

print("count of unique user:",len(unique_users))

    
```

```python
total_event_count=sum(event_count.values())


print("Sum of event counts:", total_event_count)
print("Length of cleaned_events:", len(cleaned_events))
print("Validation correct:", total_event_count == len(cleaned_events))
```

```python

csv_lines=["metric,key,value"]

for etype,count in event_count.items():
    csv_lines.append(f"count,{etype},{count}")

for etype,duration in average_duration.items():
    csv_lines.append(f"Average duration minutes,{etype},{duration}")

csv_lines.append(f"unique_users,all,{len(unique_users)}")

csv_string = "\n".join(csv_lines)


print("\nCSV-formatted string:")
print(csv_string)
    
```

```python

with open("event_summary.csv", "w") as f:
    f.write(csv_string)

with open("event_summary.csv", "r") as f:
    csv_from_file = f.read()

for line in csv_from_file.split("\n"):
    if line.startswith("count,login"):
        print("\nConfirmed from file:", line)
```
